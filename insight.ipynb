```python
import pandas as pd
```

```python
# Carregamento de Dados.
df_datainsight = pd.read_csv(r'C:\Users\User\Documents\Dados Codigo Certo\insight-basico-python\data\dados.csv', encoding='latin-1', sep=',', low_memory=False)
df_datainsight
```

```python
# Esibindo dados esstatisticos basicos
df_datainsight.describe()
```

```python
# Exibindo estatisticas da coluna Preço de venda

# Exibir a média
mean_values = df_datainsight['Preço Unitário'].mean()

# Exibir a mediana
median_values = df_datainsight['Preço Unitário'].median()

# Exibir mínimo e máximo manualmente
min_values = df_datainsight['Preço Unitário'].min()
max_values = df_datainsight['Preço Unitário'].max()

# Exibir desvio padrão
std_dev = df_datainsight['Preço Unitário'].std()

print(f'Media dos preços do curso: {mean_values} ')
print(f'Mediana dos preços do curso: {median_values} ')
print(f'Valor minimo do curso: {min_values} ')
print(f'Valor máximo do curso: {max_values} ')
print(f'Desvio padrão dos preços: {std_dev} ')
```

```python
#Exibindo estatisticas da coluna Quantidade vendida

# Exibir a média
mean_values = df_datainsight['Quantidade de Vendas'].mean()

# Exibir a mediana
median_values = df_datainsight['Quantidade de Vendas'].median()

# Exibir mínimo e máximo manualmente
min_values = df_datainsight['Quantidade de Vendas'].min()
max_values = df_datainsight['Quantidade de Vendas'].max()

# Exibir desvio padrão
std_dev = df_datainsight['Quantidade de Vendas'].std()

print(f'Quantidade media do curso: {mean_values} ')
print(f'Mediana do curso: {median_values} ')
print(f'Quantidade minima do curso: {min_values} ')
print(f'Quantidade máxima do curso: {max_values} ')
print(f'Desvio padrão da quantidade: {std_dev} ')
```

```python
# Calcular a receita total gerada pela venda dos cursos.

df_qtd_vendas = df_datainsight['Quantidade de Vendas'].sum()
df_qtd_unit = df_datainsight['Preço Unitário'].sum()

df_receita = df_qtd_vendas * df_qtd_unit


print(f'Receita TOTAL das vendas: R$ {df_receita:.2f}')
```

```python
# Identificar o curso com o maior número de vendas.

indice_max = df_datainsight['Quantidade de Vendas'].idxmax()


nome_max = df_datainsight.loc[indice_max, 'Nome do Curso']
vendas_max = df_datainsight.loc[indice_max, 'Quantidade de Vendas']

print(f'O curso: {nome_max}, teve {vendas_max} vendas no periodo.')
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Visualizar a distribuição das vendas ao longo do tempo através de gráficos.

plt.figure(figsize=(10, 6))
sns.barplot(x='Data', y='Quantidade de Vendas', data=df_datainsight, palette='viridis')

# Configurando os titulos do grafico e do eixo
plt.title('Quantidade de Vendas por Data')
plt.xlabel('Data')
plt.ylabel('Quantidade de Vendas')

# Plotando o grafico
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
# Incluindo uma coluna com a receita total dos cursos vendidos

df_datainsight['Receita Total'] = df_datainsight['Quantidade de Vendas'] * df_datainsight['Preço Unitário']
```

```python
# Grafico com a Receita Total dos cursos vendidos

# Somando as vendas dos cursos
df_datainsight.groupby('Nome do Curso').sum()

# Ordenando o curso por maior receita
df_sorted = df_datainsight.groupby('Nome do Curso').agg({'Receita Total': 'sum'}).reset_index()
df_sorted = df_sorted.sort_values(by='Receita Total', ascending=False)


#Plotando o gráfico de barras para demosntrar a receita total dos cursos
plt.figure(figsize=(12, 8))  
sns.barplot(x='Receita Total', y='Nome do Curso', data=df_sorted, color='blue')

# Configurando os titulos do grafico e do eixo
plt.title('Receita por Curso')
plt.xlabel('Receita Total')
plt.ylabel('Nome do Curso')

# Plotando o grafico
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
